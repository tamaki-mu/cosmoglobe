# gain

## in

```python
g_in = np.loadtxt('g_in.txt')
plt.plot(g_in)
plt.xlim(0,10957)
plt.title('gain_in')
```

## out

```python
g_out = []
for k in range(num_sample):
    g_out.append(np.loadtxt('g_out_00'+str(str(k).zfill(4))+'.txt'))
```

```python
np.array(g_out).shape
```

```python
g_out_mean = np.mean(g_out, axis=0)
```

```python
g_out_mean.shape
```

```python
plt.plot(g_out_mean)
plt.xlim(0,10957)
plt.title('gain_out_mean')
```

```python
g_out_sq = np.array(g_out)**2
g_out_rms = np.sqrt(np.mean(g_out_sq, axis=0))
```

```python
plt.plot(g_out_rms)
plt.xlim(0,10957)
plt.title('gain_out_rms')
```

## (mean-in)/rms

```python
g_result = (g_out_mean - g_in)/g_out_rms
```

```python
plt.plot(g_result)
plt.xlim(0,10957)
plt.title('(mean-input)/rms')
```
